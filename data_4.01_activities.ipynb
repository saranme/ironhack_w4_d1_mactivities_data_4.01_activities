```python
df.HOMEOWNR.value_counts()
```

```python
df[df.HOMEOWNR == ' ']
```

```python
def clean_home_owner(x):
    if x == ' ':
        x = 'U'
    return x

df.HOMEOWNR = list(map(clean_home_owner, df.HOMEOWNR))
```

# 4.01 Activity 2
Activity: Using linear regression to impute missing values.
You already know how to predict a numerical ammount. Therefore, you can use other columns in order to predict the missing values of the column of you interest. Use 'HV1' and 'IC1' columns to predict the missing values of 'INCOME'.

Hint: For sake of simplicity, when you have NaN s, you work with them as if they were a test set.

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression

X = df[df.INCOME.notnull()][['HV1','IC1']]
y = df[df.INCOME.notnull()]['INCOME']
X_nulls = df[df.INCOME.isnull()][['HV1', 'IC1']]
```

```python
model = LinearRegression().fit(X,y)
income_pred = model.predict(X_nulls)
df.loc[df.INCOME.isnull(), 'INCOME'] = income_pred
df.head()
```

```python
df.INCOME.value_counts()
```

# 4.01 Activity 4

- What is the difference between **map**, **filter**, and **reduce**?
- Use **map**, **filter** and **reduce** to get the sum of the square root of the odd numbers between 0 and 100.


```python
def odd_or_not(x):
    if x%2 != 0:
        return x
    else:
        return 0
sum(list(map(odd_or_not,list(range(100)))))
```

```python
import math
sum(list(filter(lambda x: math.sqrt(x) if x%2!=0 else 0, [ i for i in range(100)])))
```
